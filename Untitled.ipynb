```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
target=pd.DataFrame(data.target,columns=["MEDV"])
target
```

```python
X=np.append(np.ones((506,1)),df,axis=1)   #adding 1 to all columns for bias term
X.shape
```

```python
y=target["MEDV"]
y.shape
n=X.shape[1] 
print(n)
```

```python
def cost_function(X,y,theta):
    m=y.size                        #no of training examples
    n=X.shape[1]                     #no of features(m+1 due to bias term)
    hypo=np.dot(X,theta)
    error=hypo-y
    square_error=error**2
    MSE=np.sum(square_error)
    J=MSE/(2*m)
    return J
```

```python
def gradient_descent(X,y,alpha=0.0001,iterations=5000):
    J_iter=np.zeros(iterations)
    m=y.size                        #no of training examples
    n=X.shape[1]
    theta=np.zeros(n)
    for i in range(iterations):
        hypo=np.dot(X,theta)
        err=hypo-y
        delta=np.dot(np.transpose(X),err)/m
        theta_change=(alpha)*delta
        theta=theta-theta_change
        J_iter[i]=cost_function(X,y,theta)
    return (theta,J_iter) 
        
        
        
        
        
```

```python
(new_theta,J_iter)=gradient_descent(X,y)
```
